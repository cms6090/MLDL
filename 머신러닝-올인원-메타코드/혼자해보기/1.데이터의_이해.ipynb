#### 2.1 결측치 확인

```python
print('항목별 na개수 : \n', data.isnull().sum())
print('전체 na개수 : ', data.isnull().sum().sum())
```

#### 2.2 결측치 제거

```python
data_cleaned = data.dropna(axis=0)
data_cleaned
```

### 3. 이상치 확인
data_cleaned의 “satisfaction_level”, “last_evaluation”, “number_project”, “average_montly_hours”, “time_spend_company” 변수들의 이상치 확인

```python
data_show = data_cleaned[['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company']]
```

```python
plt.figure(figsize=(5, 5))
for idx, column in enumerate(data_show.columns):
    plt.boxplot(data_show[column])
    plt.xticks([idx],[column])
    idx += 1
    plt.show()
```

### 4. 중복데이터 다루기
data_cleaned에 대해서 중복 데이터 제거 후 data_cleaned2로 저장

```python
data_cleaned2 = data_cleaned.drop_duplicates()
data_cleaned2.shape
```

### 5. 데이터 메니퓰레이션

1. data_cleaned2 기준, last_evaluation > 0.8인 직원들 찾기

```python
data_cleaned2[data_cleaned2['last_evaluation'] > 0.8]
```

2. data_cleaned2 기준, last_evaluation > 0.8이면서 left가 1인 직원들 찾기

```python
data_cleaned2[(data_cleaned2['last_evaluation'] > 0.8) & (data_cleaned2['left'] == 1)]
```

3. data_cleaned2 기준,  “satisfaction_level”, “last_evaluation”, “number_project”, “average_montly_hours”, “time_spend_company” 변수만 포함하는 새로운 데이터 프레임을 만들어 data_cleaned_numeric에 저장

```python
data_cleaned_numeric = data_cleaned2[['satisfaction_level','last_evaluation','number_project','average_montly_hours','time_spend_company']]
data_cleaned_numeric
```

4. data_cleaned_numeric 기준, last_evaluation > 0.8이면서 left가 1인 경우에 대해 그룹핑하고 평균으로 데이터 요약하여 간단하게 해석해보기

```python
# data_cleaned2에서 left가 1인 경우를 필터링
filtered_numeric = data_cleaned2[(data_cleaned2['left'] == 1) & (data_cleaned2['last_evaluation'] > 0.8)]

# 그룹핑하고 평균 계산하기
filtered_numeric.groupby('left')[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company']].mean()
```
